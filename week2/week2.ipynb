```python
def listsum(numbers):
    """Return the sum of a list of numbers."""
    return sum(numbers)
```

```python
def deduplicate(items):
    """Return a list with duplicates removed."""
    seen = set()
    result = []
    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result
```

```python
def sorttuples(pairs):
    """Return a list of tuples sorted by their last element."""
    return sorted(pairs, key=lambda t: t[-1])
```

```python
def squarecubes(numbers):
    """Return lists of squares and cubes for each number."""
    squares = [n ** 2 for n in numbers]
    cubes = [n ** 3 for n in numbers]
    return squares, cubes
```

```python
# Tests for squarecubes
s, c = squarecubes([1, 2, 3, 4])
assert s == [1, 4, 9, 16]
assert c == [1, 8, 27, 64]

s, c = squarecubes([])
assert s == [] and c == []

s, c = squarecubes([0, -1])
assert s == [0, 1]
assert c == [0, -1]

print("All squarecubes tests passed.")
```
